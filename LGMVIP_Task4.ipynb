# **TASK-4- Image to Pencil Sketch with Python:**

*Dataset link: https://thecleverprogrammer.com/2020/09/30/pencil-sketch-with-python/*

**Importing Modules**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Read the image**

```python
image=cv2.imread("/Nature.jpg")
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

**Step 1: Convert to Grey Image**

```python
gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.axis(False)
plt.show()
```

**Step 2: Invert ImageStep**

```python
inverted_image=cv2.bitwise_not(gray_image)
plt.imshow(inverted_image)
plt.axis(False)
plt.show()
```

**Step 3: Blur image**

```python
blured_image=cv2.GaussianBlur(inverted_image, (21,21),0)
plt.imshow(blured_image)
plt.axis(False)
plt.show()
```

**Step 4: Invert Blurred Image**

```python
invblur_image=cv2.bitwise_not(blured_image)
plt.imshow(invblur_image)
plt.axis(False)
plt.show()
```

**Step 5: Sketch**

```python
sketch_image = cv2.divide(gray_image,invblur_image, scale=256.0)
plt.imshow(sketch_image)
plt.axis(False)
plt.show()

cv2.imwrite('sketch.png', sketch_image)
```

**Step 6: Display sketch**

```python
rgb_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

**Original Image vs Sketch**

```python
# ORIGINAL IMAGE
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original Image', size=30)
plt.imshow(RGB_img)
plt.axis('off')

# SKETCH IMAGE
plt.subplot(1,2,2)
plt.title('Sketch', size=30)
rgb_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
